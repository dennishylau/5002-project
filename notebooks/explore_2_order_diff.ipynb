# Find anomaly in a single time series using Second-order differencing

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.options.plotting.backend = "plotly"

BASE_PATH = '../data-sets/KDD-Cup/data/'
```

```python
df = pd.read_csv(BASE_PATH + '005_UCR_Anomaly_4000.txt', names=['series'])
fig = plt.Figure(figsize=(500,50), dpi= 100)
display(df.plot())
```

# Second Order Difference

```python
df_t1 = df.copy()
df_t1.index -= 1
df_t2 = df_t1.copy()
df_t2.index -= 1
```

```python
df_did = df.join(df_t1, rsuffix='_t1').join(df_t2, rsuffix='_t2')
df_did['series_did1'] = df_did.series_t1 - df_did.series
df_did['series_did2'] = df_did.series_t2 - 2 * df_did.series_t1 + df_did.series
df_did.head()
```

```python
df_did[['series', 'series_did1']].plot()
```

```python
df_did[['series', 'series_did2']].plot()
```
